```python
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split,KFold,StratifiedKFold
from sklearn.metrics import accuracy_score,f1_score
```

```python
X = df.drop("Crop_Damage",axis=1).values
y = df['Crop_Damage'].values
```

```python
scores = []
kfold = KFold(n_splits=5,shuffle=True,random_state=23)

for train,test in kfold.split(X):
    x_train,x_test = X[train],X[test]
    y_train,y_test = y[train],y[test]
    model = CatBoostClassifier(n_estimators=1000,random_state=23,max_depth=6)
    model.fit(x_train,y_train)
    preds = model.predict(x_test)
    score = accuracy_score(preds,y_test)
    scores.append(score)
    print(score)

print("Average score is:",sum(scores)/len(scores))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
scores = []
kfold = KFold(n_splits=5,shuffle=True,random_state=23)

for train,test in kfold.split(X):
    x_train,x_test = X[train],X[test]
    y_train,y_test = y[train],y[test]
    model = RandomForestClassifier(max_depth=6,n_estimators=400,random_state=23)
    model.fit(x_train,y_train)
    preds = model.predict(x_test)
    score = accuracy_score(preds,y_test)
    scores.append(score)
    print(score)

print("Average score is:",sum(scores)/len(scores))
```
